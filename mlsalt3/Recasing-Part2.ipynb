### Word-based Uni-gram model - KFTT-dev

```python
cseq2wseq = fst.Fst.read('cseq2wseq.fst')
lm_word_1 = fst.Fst.read('lm.word.1.fst')
```

```python
rawhyps = []
for id in xrange(1,1114):
    ifst = fst.Fst.read(KFTT_DEVINPUT+str(id)+'.fst') # read input FSA with lower case character sequences
    x1 = fst.compose(ifst, toUpper).project(project_output=True) #project to mixed case character sequences and project onto the output
    x2 = fst.compose(x1, cseq2wseq) # compose with the chara83.01cter sequence to word sequence transducer
    ofst = fst.compose(x2, lm_word_1) # compose with the unigram language model
    rawhyps.append( utilfst.printstrings(ofst, syms=chars_ST)) # get the best mixed case character sequence under the unigram language model 

chyps = [l[0].replace("<s>","").replace("</s>","") for l in rawhyps]
whyps = [line.replace(" ","").replace("_"," ") for line in chyps]

# Load the KFTT-dev character and word references:
with open(KFTT_DEVCHARS) as f:
    kfttcrefs = [line.strip() for line in f]    
with open(KFTT_DEVWORDS) as f:
    kfttwrefs = [line.strip() for line in f]
    
print 'CHARACTER-LEVEL', utilfst.evalrecasing(kfttcrefs, chyps) # character-level accuracy
print 'WORD-LEVEL', utilfst.evalrecasing(kfttwrefs, whyps) # word-level accuracy
```

```python
def num_arcs(f):
    return sum(f.num_arcs(s) for s in f.states())
```

### Character Lattice - KFTT-dev

```python
! mkdir -p prune.kfttdev.lats.lm.char.2/

for id in xrange(1,1114):
    ifst = fst.Fst.read(KFTT_DEVINPUT+str(id)+'.fst') # read input FSA with lower case character sequences
    # project to mixed case character sequences, apply the character bigram LM, and project onto the output
    x0 = fst.compose(ifst, toUpper_lm_char_2).project(project_output=True)
    x1 = fst.prune(x0, weight=10)
    fst.determinize(x1.rmepsilon()).minimize().arcsort().write('prune.kfttdev.lats.lm.char.2/'+str(id)+'.fst')
```
